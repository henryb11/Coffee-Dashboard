```python
import pandas as pd
import re
import numpy as np
```

```python
coffee = pd.read_excel('df_arabica_clean.xlsx')
```

```python
coffee.head()
```

```python
coffee = coffee.drop('Unnamed: 0', axis = 1)
```

```python
coffee.columns = coffee.columns.str.lower()
coffee.rename(columns={'country of origin': 'country_of_origin', 'farm name': 'farm_name', 
                   'lot number': 'lot_number', 'ico number': 'ico_number', 'number of bags': 'number_of_bags', 
                  'bag weight': 'bag_weight', 'in-country partner': 'in_country_partner', 
                   'harvest year': 'harvest_year', 'grading date': 'grading_date', 
                   'processing method': 'processing_method', 'clean cup': 'clean_cup', 
                   'total cup points': 'total_cup_points', 'moisture percentage': 'moisture_percentage', 
                  'category one defects': 'category_one_defects', 'category two defects': 'category_two_defects', 
                  'certification body': 'certification_body', 'certification address': 'certification_address', 
                  'certification contact': 'certification_contact'}, inplace=True)
```

```python
print(coffee.columns)
```

```python
coffee = coffee.drop(['mill','ico_number','producer','in_country_partner','owner','status','defects','quakers','certification_body','certification_address','certification_contact'],axis=1)
```

```python
coffee.head()
```

```python
def upd_alt(row):

    alt = row['altitude']
    try: 
        alt = int(alt)
        return alt 
    except: 
        try: 
            list = re.findall(r'\d+', alt)
            value_1 = int(list[0])
            value_2 = int(list[-1])
            return int((value_1 + value_2) / 2)
        except:
            return 'na'
        
```

```python
coffee['altitude'] = coffee.apply(upd_alt, axis=1)
```

```python
coffee[coffee['altitude']=='na']
```

```python
coffee.drop(coffee[coffee.altitude == 'na'].index, inplace=True)
```

```python
def gross_mass (row):
    
    bag_mass = row['bag_weight']
    bag_mass = re.search(r'\d+',bag_mass).group()
    
    num_bags = int(row['number_of_bags'])
    
    return int(bag_mass)*num_bags
```

```python
coffee['total_mass'] = coffee.apply(gross_mass, axis=1)
```

```python
coffee.groupby(by=["country_of_origin"]).sum()[['total_mass']].sort_values(by='total_mass', ascending=False)
```

```python
coffee.groupby(by=["country_of_origin"]).mean()[['total_cup_points']].sort_values(by='total_cup_points', ascending=False)
```

```python
top_8 = ['Caturra', 'Gesha', 'Typica', 'Bourbon', 'Catuai', 'unknown', 'Catimor','Ethiopian Heirlooms']

coffee[coffee['variety'].isin(top_8)].groupby(by=["variety"]).mean()[['total_cup_points']].sort_values(by='total_cup_points', ascending=False)


```
